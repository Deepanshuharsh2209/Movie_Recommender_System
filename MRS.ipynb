```python
def tag(l):
    l2=[]
    for i in ast.literal_eval(l):
        l2.append(i['name'])
    return l2
```

```python
movies['genres']=movies['genres'].apply(tag)
```

```python
def cast_tag(l):
    l2=[]
    flag=0
    for i in ast.literal_eval(l):
        if flag !=5:
            l2.append(i['name'])
            flag+=1
        else:
            break
    return l2
```

```python
movies['cast']=movies['cast'].apply(cast_tag)
```

```python
movies['crew'][0]
```

In the crew section, I want to extract the name of the director.

For that I have to look for job key as Director and then from name key extract the name of the director.

```python
def director(l):
    l2=[]
    for i in ast.literal_eval(l):
        if i['job']=='Director':
            l2.append(i['name'])
            break
    return l2
```

```python
movies['crew']=movies['crew'].apply(director)
```

```python
 movies['genres']=movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
        
```

```python
movf['tags']=movf['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(movf['tags']).toarray()
```

```python
from sklearn .metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    mov_index=movf[movf['title']==movie].index[0]
    distances=similarity[mov_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(movf.iloc[i[0]].title)
    
```

```python
#logic for above function
# I want index of movie passed in function from my dataframe, so that I can pass in in similarity, then I will recommend top 5 similar movies based on minimum cosine distance.
movf[movf['title']=='Avatar'].index[0]
```
